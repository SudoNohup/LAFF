## By columns

Consider $ C := A B + C $.  Partition
$$
C \rightarrow \left( \begin{array}{c | c | c | c }
c_0 & c_1 & \cdots & c_{n-1}
\end{array}
\right)
\mbox{ and }
B \rightarrow \left( \begin{array}{c | c | c | c }
b_0 & b_1 & \cdots & b_{n-1}
\end{array}
\right).
$$
Then
\begin{eqnarray*}
\left( \begin{array}{c | c | c | c }
c_0 & c_1 & \cdots & c_{n-1}
\end{array}
\right)
&:=&
A \left( \begin{array}{c | c | c | c }
b_0 & b_1 & \cdots & b_{n-1}
\end{array}
\right)
+ 
\left( \begin{array}{c | c | c | c }
c_0 & c_1 & \cdots & c_{n-1}
\end{array}
\right)\\
&=&
\left( \begin{array}{c | c | c | c }
A b_0 + c_0 & A b_1 + c_1 & \cdots & A b_{n-1} c_{n-1}
\end{array}
\right).
\end{eqnarray*}
So, $ C $ can by updated one column at a time: 
$ c_j := A b_j + c_j $.

## By rows

Consider $ C := A B + C $.  Partition
$$
C \rightarrow \left( \begin{array}{c}
\widehat c_0^T \\
\widehat c_1^T \\
\vdots \\
\widehat c_m-1^T
\end{array}
\right)
\mbox{ and }
A \rightarrow \left( \begin{array}{c}
\widehat a_0^T \\
\widehat a_1^T \\
\vdots \\
\widehat a_m-1^T
\end{array}
\right).
$$
Then
\begin{eqnarray*}
\left( \begin{array}{c}
\widehat c_0^T \\
\widehat c_1^T \\
\vdots \\
\widehat c_m-1^T
\end{array}
\right)
& := & 
\left( \begin{array}{c}
\widehat a_0^T \\
\widehat a_1^T \\
\vdots \\
\widehat a_m-1^T
\end{array}
\right) B + \left( \begin{array}{c}
\widehat c_0^T \\
\widehat c_1^T \\
\vdots \\
\widehat c_m-1^T
\end{array}
\right) \\
& = &
\left( \begin{array}{c}
\widehat a_0^T B + \widehat c_0^T \\
\widehat a_1^T B + \widehat c_1^T \\
\vdots \\
\widehat a_{m-1}^T B + \widehat c_{m-1}^T
\end{array}
\right)
\end{eqnarray*}
So, $ C $ can by updated one row at a time: 
$ \widehat c_i^T := \widehat a_i^T B + \widehat c_j^T $.

But notice that if you transpose both sides:
\begin{eqnarray*}
\widehat c_1 = ( \widehat c_1^T )^T &=& ( \widehat a_1^T B + \widehat c_1^T )^T \\
&=& ( \widehat a_1^T B )^T + (\widehat c_1^T)^T \\
&=& B^T ( \widehat a_1^T )^T + \widehat c_1 \\
&=& B^T \widehat a_1  + \widehat c_1.
\end{eqnarray*}

## By rank-1 updates

Consider $ C := A B + C $.  Partition
$$
A \rightarrow \left( \begin{array}{ c | c | c | c}
a_0 & a_1 & \cdots & a_{k-1}
\end{array} \right)
\mbox{ and }
B \rightarrow \left( \begin{array}{c}
\widehat b_0^T \\ \hline
\widehat b_1^T \\ \hline
\vdots \\ \hline
\widehat b_m-1^T 
\end{array}
\right)
$$
Then
\begin{eqnarray*}
C & = & 
\left( \begin{array}{ c | c | c | c}
a_0 & a_1 & \cdots & a_{k-1}
\end{array} \right)
\left( \begin{array}{c}
\widehat b_0^T \\ \hline
\widehat b_1^T \\ \hline
\vdots \\ \hline
\widehat b_m-1^T 
\end{array}
\right) + C
\\
& = &
a_0 \widehat b_0^T + 
a_1 \widehat b_1^T + 
\cdots
a_{p-1} \widehat b_{p-1}^T + C\\
& = &
a_{p-1} \widehat b_{p-1}^T + (\cdots + (a_1 \widehat b_1^T + ( a_0 \widehat b_0^T + C ) ) \cdots )
.
\end{eqnarray*}
So, $ C $ can by by repeated adding a rank-1 update to it.  (Recall that $ y x^T + A $ is called a rank-1 update.  Here $ a_p $ takes the place of $ y $, $ \widehat b_p^T $ takes the place of $ x^T $, and $ C $ takes the place of $ C $.)
